# 感知机的学习和使用

## 1. 引入外部库函数

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. 与门电路的使用尝试


```python
def AND(x1 , x2 ):
    X = np.array([x1 , x2])
    W = np.array([0.5 , 0.5])
    B = -0.7
    res = np.sum(W*X)+B
    if(res<=0):
        return 0
    else:
        return 1
AND(1 , 1),AND(0 , 1),AND(1 , 0),AND(0 , 0)
```

### 2.与非门电路的实现

```python
def AND_N(x1 , x2):
    X = np.array([x1 , x2])
    W = np.array([-0.5 , -0.5])
    B = 0.7
    res = np.sum(W*X)+B
    if res>=0:
        return 1
    else:
        return 0
AND_N(1 , 1),AND_N(0 , 1),AND_N(1 , 0),AND_N(0 , 0)
```

### 3.或门电路的实现

```python
def OR(x1 , x2):
    T1 = AND_N(x1 , x1)
    T2 = AND_N(x2 , x2)
    res = AND_N(T1 , T2)
    return res
OR(1 , 1),OR(0 , 1),OR(1 , 0),OR(0 , 0)
```

### 4.或非门电路的实现

```python
def OR_N(x1 , x2):
    T1 = AND_N(x1 , x1)
    T2 = AND_N(x2 , x2)
    res = AND(T1 , T2)
    return res
OR_N(1 , 1),OR_N(0 , 1),OR_N(1 , 0),OR_N(0 , 0)  
```

### 5.异或门电路的实现

```python
def XOR(x1 , x2):
    s1 = AND_N(x1 , x2)
    s2 = OR(x1 , x2)
    res = AND(s1 , s2)
    return res
XOR(1 , 1),XOR(0 , 1),XOR(1 , 0),XOR(0 , 0)  
```
