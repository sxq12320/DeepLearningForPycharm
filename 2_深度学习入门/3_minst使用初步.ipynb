```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os 
import torch
import gzip
from PIL import Image
```

```python
def load_mnist_from_raw(data_dir="data/MNIST/raw"):
    train_images_path = os.path.join(data_dir, "train-images-idx3-ubyte.gz")
    train_labels_path = os.path.join(data_dir, "train-labels-idx1-ubyte.gz")
    test_images_path = os.path.join(data_dir, "t10k-images-idx3-ubyte.gz")
    test_labels_path = os.path.join(data_dir, "t10k-labels-idx1-ubyte.gz")

    def _load_imgs(path):
        with gzip.open(path, 'rb') as f:
            imgs = np.frombuffer(f.read(), dtype=np.uint8, offset=16)
        return imgs.reshape(-1, 28, 28)

    def _load_lbls(path):
        with gzip.open(path, 'rb') as f:
            lbls = np.frombuffer(f.read(), dtype=np.uint8, offset=8)
        return lbls

    x_train = _load_imgs(train_images_path)
    t_train = _load_lbls(train_labels_path)
    x_test = _load_imgs(test_images_path)
    t_test = _load_lbls(test_labels_path)

    return (x_train, t_train), (x_test, t_test)
```

```python
(x_train, t_train), (x_test, t_test) = load_mnist_from_raw()

print("训练集图像形状:", x_train.shape)  
print("训练集标签形状:", t_train.shape)  
print("测试集图像形状:", x_test.shape)    
print("测试集标签形状:", t_test.shape)   
```

```python
def image_show(img):
    pil_img = Image.fromarray(np.uint8(img))
    pil_img.show()
```
